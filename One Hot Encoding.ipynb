```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df.head(2)
```

```python
#coverting categorical data to a dummy variable meaning into binary distribution
#using pandas to create a dummy variable 
dummy=pd.get_dummies(df["town"],dtype = int)
```

```python
dummy
```

```python
#To merge the dummy and initial data together
merge=pd.concat([df,dummy],axis ="columns")
merge
```

```python
#Dropping the town column
processed_data =merge.drop(["town"],axis=1)
processed_data
```

```python
final_data = processed_data.drop(["west windsor"],axis = 1)
final_data
```

```python
x= final_data.drop("price",axis=1)
x
```

```python
y=final_data["price"]
y
```

```python
model=LinearRegression()
model
```

```python
model.fit(x,y)
```

```python
#checking for the accuracy
model.score(x,y)
```

```python
model.predict(x)
```

```python
#predict 2800 sqr feet home in robinsville view
model.predict([[2800,0,1]])
```

```python
#predict 3400 sqr feet home in windsor
model.predict([[3400,0,0]])
```

```python
#predict 3400 sqr feet home in monroe township
model.predict([[3400,1,0]])
```
